# Random forest

Concepts: 
-  It consists of a large number of individual decision trees that operate as an **ensemble**.
-  Each individual tree in the random forest spits out a class prediction and the class with the most votes becomes our model’s prediction
-  A large number of relatively **uncorrelated** models (trees) operating as a committee will outperform any of the individual constituent models.
-  **Bagging** (bootstrap aggregation): each individual tree is randomly sampled from the dataset with **replacement**
-  **Feature Randomness**: each tree in a random forest can pick only from a **random** subset of features

The key benefit of RF in comparisoin to DT is that DT run the risk of overfitting as they tend to tightly fit all the samples within training data. However, when there’s a robust number of decision trees in a random forest, the classifier won’t overfit the model since the averaging of uncorrelated trees lowers the overall variance and prediction error.

# Gradient boosting 

Concepts: 
- It's an algorithm that attempts to accurately predict a target by combining the estimates of a set of simpler and weaker learners into a complex one
- Sequencial in nature - it's called as such because of it uses **gradient descent** to minimise the loss function when adding new learners to the ensemble
- It starts off with a naive decision tree (i.e. C4.5 or CART) and make some predictions
- Then it becomes an iterative process to
  - calculate the loss
  - train a new tree on the residuals (the difference between the label and prediction)
  - add the new tree to the ensemble
  - make predictions again
- The model improves as we move each tree more in the direction via small updates based on the gradient of the loss function
- The idea is that the subsequent learners are trying to predict the error of the previous learner. So we are sequencially correcting our errors.
- It usually outperforms RF

XGBoost provides a parallel tree boosting (also known as GBDT, GBM) that solve many data science problems in a fast and accurate way.
