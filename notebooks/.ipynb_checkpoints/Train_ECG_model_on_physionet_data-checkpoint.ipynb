```python
import logging 

import matplotlib.pyplot as plt
import numpy as np
import sklearn.model_selection

import algo_ecg.data
import algo_ecg.preprocessing
#import algo_ecg.model
```

```python
logger = logging.getLogger()
logger.setLevel(logging.DEBUG)

plt.rcParams['figure.figsize'] = [12, 8]
```

```python
X_in, y = algo_ecg.data.import_physionet_data('../data', num_files_to_read=100)
```

```python
X = algo_ecg.preprocessing.preprocess(X_in)
```

```python
plt.plot(X_in[71])
plt.plot(X[71])
```

```python
model = algo_ecg.model.create_model(dropout=.2, lstm_units=8)
```

```python
model.summary()
```

```python
X_train, X_test, y_train, y_test = sklearn.model_selection.train_test_split(X, y, train_size=.7)
```

```python
slice_length = len(X_train[0])
X_train = np.concatenate(X_train).reshape(-1, slice_length, 1)
X_test = np.concatenate(X_test).reshape(-1, slice_length, 1)
```

```python
%%time 
model.fit(
    x=X_train, 
    y=y_train, 
    batch_size=1024,
    epochs=2000
)
```

```python
model.save('rcnn_ecg_afib_v_normal_30sec_slices.hdf5')
```

```python
%%time 
pred = model.predict_proba(X_test)
```

```python
fpr, tpr, _ = sklearn.metrics.roc_curve(y_test, pred)
```

```python
sklearn.metrics.auc(x=fpr, y=tpr)
```

```python
plt.plot(fpr, tpr)
```

```python
CUTOFF = .5
acc = sklearn.metrics.accuracy_score(y_true=y_test, y_pred=(pred > CUTOFF))
prec = sklearn.metrics.precision_score(y_true=y_test, y_pred=(pred > CUTOFF))
rec = np.sum(y_test * (pred > CUTOFF).reshape((-1,))) / np.sum(y_test)
sens = np.sum((y_test + (pred > CUTOFF).reshape((-1,))) == 0) / np.sum(y_test == 0)
acc, prec, rec, sens
```
